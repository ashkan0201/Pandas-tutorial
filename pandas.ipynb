```python
import pandas as pd
```

```python
# The left part is our indexes and the right part is our data.

# My Data
data = ["A", "s", "h", "k", "a", "n"]

result = pd.Series(data)

print(result)
```

```python
# Defining an arbitrary index on the data.

# My Data
data = ["A", "s", "h", "k", "a", "n"]

result = pd.Series(data, index=[1, 3, 5, 7, 9, 11])

print(result)
print("\n")
print(result.values)
print(result.index)
```

```python
# Indexing in the list.
# In this section, if you define an index for your data, when you get the index,
#you must give exactly the index of the same values that you defined for the index in the serial argument.

# My Data
data = ["A", "s", "h", "k", "a", "n"]

result = pd.Series(data, index=["in0", "in1", "in2", "in3", "in4", "in5"])

print(result["in0"], result["in1"], result["in2"], result["in3"], result["in4"], result["in5"])
print(result[0], result[1], result[2], result[3], result[4], result[5])
```

```python
# Slicing with Pandaz functions.

# My Data
data = [10, 20, 30, 40, 50, 60]

result = pd.Series(data, index=["ten", "twenty", "thirty", "forty", "fifty", "sixty"])

# Indexing with the iloc function means to get the index from the data itself
print(result.iloc[:4])

print("\n")

# Indexing with the loc function means to take from the defined index itself
print(result.loc["ten":"forty"])
```

```python
# Slicing

# My Data
data = [10, 20, 30, 40, 50, 60]

result = pd.Series(data, index=["ten", "twenty", "thirty", "forty", "fifty", "sixty"])
print(result["ten":"forty"])
print("\n")
print(result[0:4])
```

```python
# In this section, the data is in the form of a dictionary.
# The key in this section is the index and the value is our data.

# My Data
data = {
    "a":"Ashkan",
    "b":"Python",
    "c":"Pandas"
}

result = pd.Series(data)
print(result)
```

```python
# Define arbitrary index
# Pay attention, when you define an index in this section,
# the index values must be equal to the dictionary key values


# My Data
data = {
    "a":"Ashkan",
    "b":"Python",
    "c":"Pandas"
}

result = pd.Series(data, index=["a", "b", "c", "d"])

# Getting data with index in two ways:

# 1
print(result["a"], result["b"], result["c"] ,result["d"])

# 2
print(result[0], result[1], result[2], result[3])
```

### Mathematical calculations in pandaz.

```python
# My Data
data1 = ["85", [85], 85.85, ("85"), 85]
data2 = ["85", [85], 85.85, ("85"), 85]
data3 = [1, 2, 3, 4]
data4 = [4, 3, 2, 1]

ser1 = pd.Series(data1)
ser2 = pd.Series(data2)
ser3 = pd.Series(data3)
ser4 = pd.Series(data4)

print("Addition")
print(ser1.add(ser2))

print("\nSubtraction")
print(ser3.sub(ser4))

print("\nDistribution")
print(ser3.div(ser4))

print("\nMultiplication")
print(ser3.mul(ser4))

print("\nPower")
print(ser3.pow(ser4))
```

```python
# Column indexing

# My Data
data = {
    "Name":["Ashkan", "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[872634, 981239, 918243]
}

result = pd.DataFrame(data)

result[["Name", "Id"]]
```

```python
# Row indexing

# My Data
data = {
    "Name":["Ashkan", "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[872634, 981239, 918243]
}

result = pd.DataFrame(data, index=["0", "1", "2"])

# Vertical
print(result.loc["0"])
print("\n")
print(result.iloc[0])

print("\n")

# Horizontal
print(result.loc[["0"]])
print("\n")
print(result.iloc[[0]])
```

```python
# Row Slicing

# My Data
data = {
    "Name":["Ashkan", "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[872634, 981239, 918243]
}

result = pd.DataFrame(data, index=["0", "1", "2"])

# 1
print(result.loc["1":"2"])

print("\n")

# 2
print(result.iloc[1:])
```

```python
# Slashing rows and columns together

# My Data
data = {
    "Name":["Ashkan", "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[872634, 981239, 918243]
}

result = pd.DataFrame(data, index=["0", "1", "2"])

# The right part of the bracket is for columns and the other part for rows

# 1
print(result.loc["0":"1", "Name":"Age"])

print("\n")

# 2
print(result.iloc[0:2, 0:2])
```

```python
# Comparative

# My Data
data = {
    "Name":["Ashkan", "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[872634, 981239, 918243]
}

result = pd.DataFrame(data, index=["0", "1", "2"])

# 1
print(result["Name"], "\n")

# 2
print(result["Name"] == "Ashkan", "\n")

# 3
print(result["Age"] > 19, "\n")
```

```python
# Query with condition.

# My Data
data = {
    "Name":["Ashkan", "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[872634, 981239, 918243]
}

result = pd.DataFrame(data, index=["0", "1", "2"])

# 1
print(result[result["Name"] == "Ashkan"], "\n")

# 2
print(result[result.index == "1"])
```

```python
# Null values

# My Data
data = {
    "Name":[None, "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[None, 981239, 918243]
}

result = pd.DataFrame(data)

# Null
print(result.isnull(), "\n")

# Not null
print(result.notnull(), "\n")

# Where are the nulls?
print(result[result.notnull()])
```

```python
# Assign values to null values.

# My Data
data = {
    "Name":[None, "Sara", "Donya"],
    "Age":[19, 20, 18],
    "Id":[None, 981239, 918243]
}

result = pd.DataFrame(data)

result.fillna("Im not null!")
```

```python
#My datas
data1 = {
    "Name":["ashkan","ali","mamad"],
    "Score":[100,39,55],
    "Quality":["High","Low","Mid"]
}
data2 = {
    "Name":["ashkan","ali","mamad"],
    "Score":[95,70,22],
    "Quality":["High","Mid","Low"]
}

result1 = pd.DataFrame(data1)
result2 = pd.DataFrame(data2)
```

```python
# Revision of additional values.
pd.merge(result1, result2, on="Name")
```

```python
# Connect two data.
result1.set_index("Name", inplace=True)
result2.set_index("Name", inplace=True)

result1.join(result2, lsuffix="_left")
```

```python
# Find identical values of a column.

#My datas
data1 = {
    "Name":["ashkan","majid","sara"],
    "Score":[100,39,55],
    "Quality":["High","Low","Mid"]
}
data2 = {
    "Name":["ashkan","ali","mamad"],
    "Score":[95,70,22],
    "Quality":["High","Mid","Low"]
}

result1 = pd.DataFrame(data1)
result2 = pd.DataFrame(data2)

final = pd.merge(result1, result2, on="Name")
print(final)
```

```python
# Participating in two data or not participating in two data.
```

```python
#left
final = pd.merge(result1, result2, on="Name", how="left")
final
```

```python
#right
final = pd.merge(result1, result2, on="Name", how="right")
final
```

```python
# Total data
final = pd.merge(result1, result2, on="Name", how="outer")
final
```

```python
# Concatenate two data
final = pd.concat([result1, result2])
final
```
